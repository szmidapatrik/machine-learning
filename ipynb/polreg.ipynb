# Polynomial regression

## Generate data

```python
import numpy as np
import matplotlib.pyplot as plt

# Polynomial function
def f(x,N):
    y = (x+9) * (x-6) * (x+5.6) + np.random.randn(N) * 1000
    return y

# Parameters
N = 50
x = np.linspace(-20,20,N)
y = f(x,N)

# Plot
plt.scatter(x,y,c='black')
```

## Model

```python
# w calculation prep
Φ_1 = np.vstack([np.ones(N), x]).T
Φ_2 = np.vstack([np.ones(N), x, np.power(x,2)]).T
Φ_3 = np.vstack([np.ones(N), x, np.power(x,2), np.power(x,3)]).T
Φ_4 = np.vstack([np.ones(N), x, np.power(x,2), np.power(x,3), np.power(x,4)]).T

# Calculate w 
w_1 = np.linalg.solve(Φ_1.T @ Φ_1, Φ_1.T @ y)
w_2 = np.linalg.solve(Φ_2.T @ Φ_2, Φ_2.T @ y)
w_3 = np.linalg.solve(Φ_3.T @ Φ_3, Φ_3.T @ y)
w_4 = np.linalg.solve(Φ_4.T @ Φ_4, Φ_4.T @ y)

# y
y_1 = Φ_1 @ w_1
y_2 = Φ_2 @ w_2
y_3 = Φ_3 @ w_3
y_4 = Φ_4 @ w_4

# Plot
plt.scatter(x,y,c='black')
plt.plot(x, y_1, c='steelblue', label='1st degree (linear)', linestyle='--')
plt.plot(x, y_2, c='purple', label='2nd degree (parabola)')
plt.plot(x, y_3, c='blue', label='3rd degree', linewidth=2.5)
plt.plot(x, y_4, c='cyan', label='4th degree', linewidth=3, linestyle=':')
plt.title('w:   b = ' + str(np.round(w_2[0], 5)) + ', m = ' + str(np.round(w_2[1], 5)))
plt.legend()
```
