# Polynomial regression

## Generate data

```python
import numpy as np
import matplotlib.pyplot as plt

# Polynomial function
def f(x,N):
    y = (x-9) * (x + 4) * (x + 5) + np.random.randn(N) * 1000
    return y

# Parameters
N = 50
x = np.linspace(-20,20,N)
y = f(x,N)

# Plot
plt.scatter(x,y)
```
