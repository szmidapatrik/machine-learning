# Polynomial regression

## Generate data

```python
import numpy as np
import matplotlib.pyplot as plt

# Polynomial function
def f(x,N):
    y = (x+9) * (x-6) * (x+5.6) + np.random.randn(N) * 1000
    return y

# Parameters
N = 50
x = np.linspace(-20,20,N)
y = f(x,N)

# Plot
plt.scatter(x,y,c='black')
```

## Model

```python
# w calculation prep
_1 = np.vstack([np.ones(N), x]).T
_2 = np.vstack([np.ones(N), x, np.power(x,2)]).T
_3 = np.vstack([np.ones(N), x, np.power(x,2), np.power(x,3)]).T
_4 = np.vstack([np.ones(N), x, np.power(x,2), np.power(x,3), np.power(x,4)]).T

# Calculate w 
w_1 = np.linalg.solve(_1.T @ _1, _1.T @ y)
w_2 = np.linalg.solve(_2.T @ _2, _2.T @ y)
w_3 = np.linalg.solve(_3.T @ _3, _3.T @ y)
w_4 = np.linalg.solve(_4.T @ _4, _4.T @ y)

# y
y_1 = _1 @ w_1
y_2 = _2 @ w_2
y_3 = _3 @ w_3
y_4 = _4 @ w_4

# Plot
plt.scatter(x,y,c='black')
plt.plot(x, y_1, c='blue', label='1nd degree polynom', linestyle='--')
plt.plot(x, y_2, c='teal', label='2nd degree polynom')
plt.plot(x, y_3, c='steelblue', label='3rd degree polynom')
plt.plot(x, y_4, c='cyan', label='4th degree polynom', linewidth=4, linestyle=':')
plt.title('w:   b = ' + str(np.round(w_2[0], 5)) + ', m = ' + str(np.round(w_2[1], 5)))
plt.legend()
```
