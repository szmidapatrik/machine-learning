```python
from ml.distributions import NormalDistribution
from ml.distributions import StudentsτDistribution
import numpy as np
import matplotlib.pyplot as plt

std = StudentsτDistribution(0,1,1)
norm = NormalDistribution(0,1)
data = norm.sample(100)
x = np.linspace(-5,5,100)
plt.scatter(data, np.zeros(100), alpha=0.2)
plt.plot(x, norm.pdf(x))
plt.plot(x, std.pdf(x))
```
