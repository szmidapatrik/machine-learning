```python
from scitools.distribution import StudentsτDistribution
from scipy.stats import t
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5,100)
std = StudentsτDistribution(0,1,1)
plt.plot(x, std.cdf(x))
plt.plot(x, t.cdf(x, 1))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def t(X, µ):
    return (np.mean(X) - µ) / (np.std(X) / np.sqrt(len(X)))


µ = 0
n = 100
data = np.random.normal(0,5,n)
mean = np.mean(data)
var = np.std(data)
res = t(data, µ)
print(res)

plt.scatter(data, np.zeros(n), alpha=0.4, zorder=2)
plt.plot(np.linspace(-20,20,n), np.zeros(n), c="black", zorder=1)
plt.plot(np.linspace(-20,20,n), norm.pdf(np.linspace(-20,20,n), 0, 5), linestyle="--", c='royalblue')
plt.plot(np.linspace(-20,20,n), norm.pdf(np.linspace(-20,20,n), np.mean(data), np.std(data)), c="slateblue")
plt.fill_between(np.linspace(-20,20,n), norm.pdf(np.linspace(-20,20,n), 0, 5), norm.pdf(np.linspace(-20,20,n), np.mean(data), np.std(data)), color="lightblue", alpha=0.8)
```
