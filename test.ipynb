```python
from ml.distributions import StudentsτDistribution
from scipy.stats import t
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5,100)
std = StudentsτDistribution(0,1,1)
plt.plot(x, std.cdf(x))
plt.plot(x, t.cdf(x, 1))
```

```python

from ml.regression import LinearRegression
import matplotlib.pyplot as plt

linreg = LinearRegression(1,1)
data = linreg.sample(100,0,5,0.5,3,0,0.5)

plt.scatter(data[:,0], data[:,1])
plt.plot(data[:,0], linreg.linear_function(data[:,0]), c="lightgrey")
plt.plot(data[:,0], linreg.fit_predict_ML(data[:,0], data[:,1]))
plt.plot(data[:,0], linreg.fit_predict(data[:,0], data[:,1]))
```

```python
from ml.classification import NaivBayes
import matplotlib.pyplot as plt

nb = NaivBayes()
X,y = nb.sample(3, 100)
fig,ax=plt.subplots(1,2,figsize=(14,6))
ax[0].scatter(X[:,0], X[:,1], c=y)
ax[1].scatter(X[:,0], X[:,1], c=nb.fit_predict(X,y))
```

```python
def polynomial_function(x, pol_params=None):
    if pol_params != None:
        if np.array(pol_params).shape == (3,):
            params = pol_params
        elif np.array(pol_params).shape == ():
            params = np.ones(3) * pol_params
        else:
            print('The given @pol_params parameter\'s type was incorrect. It must either be an integer, float, or an array of numbers with the length of 3.\nA random polynom will be generated.')
            params = np.random.randint(-10,10,3)
    else:
        params = np.random.randint(-10,10,3)
    return (x + params[0]) * (x + params[1]) * (x + params[2])

x = np.linspace(0,5,100)
plt.plot(x, polynomial_function(x,[-4, -3,  0]))
```
