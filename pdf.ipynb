# Data generation

```python
import numpy as np
import matplotlib.pyplot as plt

# Normal distribution function
def normal_pdf(X, μ, σ_2):
    σ = np.sqrt(σ_2)
    return (1 / σ * np.sqrt(2 * np.pi)) * np.exp( -(1/2) * (((X - μ) / σ) ** 2) )

# Normal distribution params
μ = 0
σ_2 = 1

# Data generation params
s = 50
s_axis = 200

# Data and axis generation
x_axis = np.linspace(-10,10, s_axis)
X = np.random.normal(μ, σ_2, s)


# Plot
fig, ax = plt.subplots(figsize=(10,6))
ax.set_title('Normal distribution PDF and CDF')
ax.scatter(X[:], np.zeros(len(X)), zorder=2)
ax.plot(x_axis, np.zeros(s_axis), c='black', zorder=1)
ax.plot(x_axis, normal_pdf(x_axis, μ, σ_2), label='N(μ = {}, σ² = {})'.format(μ, σ_2))
ax.annotate('x', (4.5,-0.009), textcoords="offset points", xytext=(0,10),ha='center')
ax.set_xlim(-5,5)
ax.legend()
```
