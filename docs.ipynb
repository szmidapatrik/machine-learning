### 1.1.1 Linear regression

The linear regression class is able to learn linear relationships between random variables and estimate them using ML (Maximum Likelihood) and MAP (Maximum A Posteriori) estimates. The Linear regression can be imported like this:
```
from scitools.regression import LinearRegression
```

Functions:

**I. linear_function(***axis*, *m*, *b***):**     *A function that creates a line from the axis array using the given* ***m*** *and* ***b*** *parameters.*



````python
from scitools.regression import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0,5,11)
y = LinearRegression().linear_function(x, 0.5, 1)

fig, ax = plt.subplots(1,2,figsize=(12,5))
ax[0].set_title('Generated datapoints (x)')
ax[0].scatter(x,np.zeros(len(x)), c='black', label='x')
ax[1].set_title('Linear function result (y)')
ax[1].scatter(x,np.zeros(len(x)), c='black', label="x")
ax[1].scatter(x,y, label="y")
ax[1].plot(x,y)
ax[0].legend()
ax[1].legend()
````
