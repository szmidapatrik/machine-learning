# Linear regression with Basis functions

## Generate data

```python
import numpy as np
import matplotlib.pyplot as plt

# Polynomial function
def f(x,N):
    y = (x+9) * (x-13) * (x+9) + np.random.randn(N) * 1000
    return y

# Parameters
N = 50
x = np.linspace(-20,20,N)
y = f(x,N)

# Plot
plt.scatter(x,y,c='black')
```

## Fixed basis points

```python
# b - base points,  - regularization
_1 = 0.5
_2 = 0.05
_3 = 0.005

b_1 = np.linspace(-20,20,5)
b_2 = np.linspace(-20,20,15)
b_3 = np.linspace(-20,20,25)

dist_1 = np.abs(x[:, np.newaxis] - b_1)
dist_2 = np.abs(x[:, np.newaxis] - b_2)
dist_3 = np.abs(x[:, np.newaxis] - b_3)

_1_1 = np.exp(-_1 * dist_1 ** 2)
_1_2 = np.exp(-_1 * dist_2 ** 2)
_1_3 = np.exp(-_1 * dist_3 ** 2)

_2_1 = np.exp(-_2 * dist_1 ** 2)
_2_2 = np.exp(-_2 * dist_2 ** 2)
_2_3 = np.exp(-_2 * dist_3 ** 2)

_3_1 = np.exp(-_3 * dist_1 ** 2)
_3_2 = np.exp(-_3 * dist_2 ** 2)
_3_3 = np.exp(-_3 * dist_3 ** 2)


# Calculate w 
w_1_1 = np.linalg.solve(_1_1.T @ _1_1, _1_1.T @ y)
w_1_2 = np.linalg.solve(_1_2.T @ _1_2, _1_2.T @ y)
w_1_3 = np.linalg.solve(_1_3.T @ _1_3, _1_3.T @ y)

w_2_1 = np.linalg.solve(_2_1.T @ _2_1, _2_1.T @ y)
w_2_2 = np.linalg.solve(_2_2.T @ _2_2, _2_2.T @ y)
w_2_3 = np.linalg.solve(_2_3.T @ _2_3, _2_3.T @ y)

w_3_1 = np.linalg.solve(_3_1.T @ _3_1, _3_1.T @ y)
w_3_2 = np.linalg.solve(_3_2.T @ _3_2, _3_2.T @ y)
w_3_3 = np.linalg.solve(_3_3.T @ _3_3, _3_3.T @ y)

# y
y_b_1 = _1_1 @ w_1_1
y_b_2 = _1_2 @ w_1_2
y_b_3 = _1_3 @ w_1_3

y_b_4 = _2_1 @ w_2_1
y_b_5 = _2_2 @ w_2_2
y_b_6 = _2_3 @ w_2_3

y_b_7 = _3_1 @ w_3_1
y_b_8 = _3_2 @ w_3_2
y_b_9 = _3_3 @ w_3_3

# Plot
fig, ax = plt.subplots(1,3,figsize=(16,5))

ax[0].scatter(x,y,c='black')
ax[0].plot(x, y_b_1, label='=0.5', linestyle='--')
ax[0].plot(x, y_b_4,c='darkslateblue', label='=0.05', linestyle="-.", linewidth=3)
ax[0].plot(x, y_b_7,c='darkturquoise', label='=0.005', linewidth=2)
ax[0].set_title('b=5')
ax[0].legend()


ax[1].scatter(x,y,c='black')
ax[1].plot(x, y_b_2, label='=0.5', linestyle='--')
ax[1].plot(x, y_b_5,c='darkslateblue', label='=0.05', linestyle="-.", linewidth=3)
ax[1].plot(x, y_b_8,c='darkturquoise', label='=0.005', linewidth=2)
ax[1].set_title('b=15')
ax[1].legend()

ax[2].scatter(x,y,c='black')
ax[2].plot(x, y_b_3, label='=0.5', linestyle='--')
ax[2].plot(x, y_b_6,c='darkslateblue', label='=0.05', linestyle="-.", linewidth=3)
ax[2].plot(x, y_b_9,c='darkturquoise', label='=0.005',linewidth=2)
ax[2].set_title('b=25')
ax[2].legend()
```
